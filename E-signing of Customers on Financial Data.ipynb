## Split the Dataset

```python
data.head()
```

```python
# Independent / Matrix of Features
x = data.drop(columns=['entry_id','e_signed'])
```

```python
# Traget
y = data['e_signed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test , y_train ,y_test = train_test_split(x , y , test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
```

```python
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Part 2: Building the Model

## 1) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier_lr = LogisticRegression(random_state=0)
```

```python
classifier_lr.fit(x_train,y_train)
```

```python
y_pred = classifier_lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

## 2) Support Vector Machines(SVM)

```python
from sklearn.svm import SVC
```

```python
classifier_svc = SVC(random_state=0)
```

```python
classifier_svc.fit(x_train,y_train)
```

```python
y_pred = classifier_svc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

## 3) Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier_rf = RandomForestClassifier(random_state=0)
classifier_rf.fit(x_train,y_train)
```

```python
y_pred = classifier_rf.predict(x_test)
```

```python
accuracy_score(y_pred , y_test)
```

```python
confusion_matrix(y_pred , y_test)
```

## 4) XGBoost

```python
from xgboost import XGBClassifier
```

```python
classifier_xgb = XGBClassifier(random_state=0)
```

```python
classifier_xgb.fit(x_train,y_train)
```

```python
y_pred = classifier_xgb.predict(x_test)
```

```python
accuracy_score(y_pred , y_test)
```

```python
confusion_matrix(y_pred , y_test)
```

```python
## Final Model - XGBoost Classifier
```

# Part 4: Applying Randomized Search to find best parameters

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters ={
    'learning_rate':[0.05,0.10,0.15,0.25,0.30],
    'max_depth':[3,4,5,6,8,10,12,15],
    'min_child_weight':[1,3,5,7],
    'gamma':[0.0,0.1,0.2,0.3,0.4],
    'colsample_bytree':[0.3,0.4,0.6,0.7],
    'n_estimators':[100,200,300,400,500],
    'subsample':[0.5,0.7,1.0] 
}
```

```python
parameters
```

```python
random_cv = RandomizedSearchCV(estimator =classifier_xgb , param_distributions=parameters , n_iter=5, n_jobs=-1,
                              scoring = 'roc_auc',cv=5 , verbose = 3)
```

```python
random_cv.fit(x_train,y_train)
```

```python
random_cv.best_estimator_
```

```python
random_cv.best_params_
```

```python
random_cv.best_score_
```

# Part 4: Final Model(XGBoost Classifier)

```python
from xgboost import XGBClassifier
classifier = XGBClassifier(base_score=0.5, booster='gbtree', callbacks=None,
              colsample_bylevel=1, colsample_bynode=1, colsample_bytree=0.5,
              early_stopping_rounds=None, enable_categorical=False,
              eval_metric=None, gamma=0.4, gpu_id=-1, grow_policy='depthwise',
              importance_type=None, interaction_constraints='',
              learning_rate=0.05, max_bin=256, max_cat_to_onehot=4,
              max_delta_step=0, max_depth=6, max_leaves=0, min_child_weight=5,
                monotone_constraints='()', n_estimators=100,
              n_jobs=0, num_parallel_tree=1, predictor='auto', random_state=0,
              reg_alpha=0, reg_lambda=1)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
accuracy_score(y_pred , y_test)
```

```python
confusion_matrix(y_pred , y_test)
```
